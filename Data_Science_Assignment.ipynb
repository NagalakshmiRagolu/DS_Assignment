**Q27: Inverse of a 2x2, then verify identity**

```python
# approach: small numeric example, then verify with allclose

A27 = np.array([[4, 7],
                [2, 6]], dtype=float)
A27_inv = np.linalg.inv(A27)
identity_check = A27 @ A27_inv
print("\nA27:\n", A27)
print("Inverse A27_inv:\n", A27_inv)
print("A27 @ A27_inv (should be identity):\n", identity_check)
print("Is close to identity", np.allclose(identity_check, np.eye(2)))
```

**Q28: Eigenvalues and eigenvectors of a 2x2**

```python
B28 = np.array([[2, 1],[1, 2]], dtype=float)
eigvals, eigvecs = np.linalg.eig(B28)
print("\nMatrix B28:\n", B28)
print("Eigenvalues:", eigvals)
print("Eigenvectors (columns):\n", eigvecs)

# verify: A v = lambda v for each eigenpair
for i, lam in enumerate(eigvals):
    lhs = B28 @ eigvecs[:, i]
    rhs = lam * eigvecs[:, i]
    print(f"Verify eigenpair {i}: close? {np.allclose(lhs, rhs)}")
```

**Q29: Solve 2x + 3y = 5 and 4x + 6y = 10**

```python
# Equations:
# 2x + 3y = 5
# 4x + 6y = 10

# Parametric solution: let y = t
def solve_equations(t):
    x = (5 - 3*t) / 2
    y = t
    return x, y

# Show a few solutions
for t in [0, 1, 2, 3]:
    print(f"t={t} -> (x, y) = {solve_equations(t)}")
```

**Q30: SVD on a 3x3 and reconstruct matrix**

```python
import numpy as np

# Step 1: Enter matrix manually
print("Enter 3x3 matrix values row by row:")
A = []
for i in range(3):
    row = list(map(float, input(f"Row {i+1}: ").split()))
    A.append(row)

A = np.array(A)
print("\nMatrix A:\n", A)

# Step 2: Do SVD
U, s, Vt = np.linalg.svd(A)

# Step 3: Build S
S = np.diag(s)

# Step 4: Reconstruct using np.dot step by step
temp = np.dot(U, S)
A_reconstructed = np.dot(temp, Vt)

print("\nReconstructed A:\n", A_reconstructed)
```
